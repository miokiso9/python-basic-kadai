```python
class Human:
    # コンストラクタを定義する,属性を定義する
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # メソッドを定義する
    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}{self.age}歳は大人です")
        else:
            print(f"{self.name}{self.age}歳は大人ではありません")

# インスタンスを複数作成し、リストに追加する
humans = [
    Human("侍一郎", 25),
    Human("侍二郎", 20),
    Human("侍花子", 16)
]

# リストの要素数分だけcheck_adultを呼び出す
for human in humans:
    human.check_adult()
```

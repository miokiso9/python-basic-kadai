```python
class Human:
    # コンストラクタを定義する,属性を定義する
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # メソッドを定義する
    def printinfo(self):
        print(self.name)
        print(self.age)

# インスタンス化する
human = Human("侍太郎", 36)

# メソッドにアクセスし、値を出力する
human.printinfo()
```
